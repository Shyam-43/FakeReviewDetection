```python
# Import necessary libraries
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import dataset
df=pd.read_csv('dataset.csv')
```

```python
# Introduction
print("### Dataset Overview ###")
print(f"Dataset Shape: {df.shape}")
print(df.head())
```

```python
# Missing Value Analysis
print("\n### Missing Values Before Cleaning ###")
print(df.isnull().sum())
```

```python
# Drop unnecessary columns
if "Unnamed: 0" in df.columns:
    df = df.drop(columns=["Unnamed: 0"])
    print("\nDropped 'Unnamed: 0' column.")
```

```python
# Drop rows with missing values in 'text_'
df = df.dropna(subset=["text_"])
print("\n### Missing Values After Cleaning ###")
print(df.isnull().sum())
```

```python
# Visualize Class Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="label", palette="viridis")
plt.title("Class Distribution")
plt.show()
```

```python
# Proportion of Ratings
rating_proportions = df['rating'].value_counts(normalize=True) * 100  
print("\n### Proportion of Ratings ###")
print(rating_proportions)

# Visualize the proportion of ratings
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="rating", palette="viridis")
plt.title("Proportion of Ratings")
plt.show()
```

```python
# Display information
df.info()
```

```python
# Data summary
df.describe()
```

```python
# Text Cleaning
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
# Text cleaning function
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation and numbers
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words])
    return text
```

```python
# Apply cleaning to 'text_' column
df['text_'] = df['text_'].apply(clean_text)
```

```python
# Final Dataset Summary
print("\n### Final Dataset ###")
print(f"Shape: {df.shape}")
print(df.head())
```

```python
# Save preprocessed dataset
cleaned_file_path = "preprocessed_dataset.csv"
df.to_csv(cleaned_file_path, index=False)

print(f"Preprocessed dataset saved to {cleaned_file_path}")
```
